##Import necessary modules

```python
import pickle

import histogram
import intervals
```

##Load data

```python
data = pickle.load(file("../data/sample-histogram.pickle"))
hist = histogram.Histogram(data[0], data[1])
```

##Get peaks by slope method and plot them

```python
hist.get_peaks(method='slope')
hist.plot()
```

##Get peaks by interval method and plot them

```python
#In the data/ folder, there is a pickle file with some example intervals, 
#in this case, just-intonation intervals for music. They can refer to any intervals!
ji_intervals = pickle.load(file('../data/ji-intervals.pickle'))
ji = intervals.Intervals(ji_intervals)

hist.get_peaks(method='interval', intervals=ji)
hist.plot(intervals=ji.intervals)
```
